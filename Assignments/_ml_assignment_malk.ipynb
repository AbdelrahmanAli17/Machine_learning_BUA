```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('employee_salary_data.csv')
print(df.head()) #only shows the first 5 rows so I can check if it loaded correctly.
x = df[["YearsExperience"]]   # X â†’ input
y = df["Salary"]  #what i want to predict
```

here i learned that the deafult test and train are 255 and 75% so we need to custmoise it to what we need by using 0.3 to represent the 30%

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(x_train, y_train) #l fit function where the model gets traind
y_pred = model.predict(x_test)
```

ig mse checks how far the guesses are from the real salaries so the smaller the number the better the model did

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

3ady what we write every section to show results w keda

```python
plt.scatter(x_test, y_test, color='blue', label='Real Data')
plt.plot(x_test, y_pred, color='red', label='Prediction Line')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.legend()
plt.show()
```

the red line almost fits the blue points, so the linear model did a good job at learning this pattern

```python
print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```
