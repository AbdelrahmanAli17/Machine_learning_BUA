```python
#import numpy , pandas , matplotlib , sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

a)	Load the iris dataset in Python using Scikit-Learn.


```python
iris = datasets.load_iris()
x = iris.data
y = iris.target

print(x.shape)
print(y.shape)
print(iris.target)
```

b)	Identify which species each target number (0, 1, 2) represents
Each species

```python
print(iris.target_names)
```

c)	Analyze the data using Pandas.


```python
#create your df
iris_df = pd.DataFrame(
    data= np.c_[iris["data"],iris["target"]],
    columns= iris["feature_names"] + ["target"]
)
print(iris_df)
print(iris_df.head(5))
#show info
print(iris_df.info())
print("--------------------------------------------------")
# show stat
print(iris_df.describe())
```

d)	Visualize the relationship between Sepal Length and Sepal Width.


```python
#make scatter plot
plt.figure(figsize = (10,7))
scatter = plt.scatter(x = iris_df["sepal length (cm)"], y = iris_df["sepal width (cm)"], c = iris_df["target"])
plt.title("The relationship between sepal length and sepal width in cm")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
#show legends for labels 
# Create legend manually
plt.legend(handles=scatter.legend_elements()[0], 
           labels=iris.target_names.tolist(),
           title="Species")
plt.show()



```

e)	Calculate the average sepal length for each of the three species.


```python
iris_df['species'] = iris_df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print(iris_df)
```

```python
Ave_sepal_length = iris_df.groupby("species")["sepal length (cm)"].mean()

print(Ave_sepal_length)
```
